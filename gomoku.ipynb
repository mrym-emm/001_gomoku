```python
# The logic begind this is to take the position intended which will be represented such as ("Number.Letter"), and place a
# stone at that position. 
# As the position is a string, we will need to parse it into integers

def place_on_board(board, stone, position):
    
    
    
    x = int(position[0])                                      
    y = position[1].upper() 
    
    # ^Here we seperate the position, for example ("0","A") to "0" and save it in x; and "A" in y.
    # ^x here refers to the row index of board and y refers to the column index of board.
    # ^As x is already a number but in string form, we just can just use the int() function.
    
    #v Below, as y is letters, we need to translate it into what we intend, hence multiple if-elif statements.
    #v This is not the most effective way, but it works. I could use dictionary, but I needed an option if user invalid inputs.
    
    
    if y == "A":
        y = 0
    elif y == "B":
        y = 1
    elif y == "C":
        y = 2
    elif y == "D":
        y = 3
    elif y == "E":
        y = 4
    elif y == "F":
        y = 5
    elif y == "G":
        y = 6
    elif y == "H":
        y = 7
    elif y == "I":
        y = 8
    elif y == "J":
        y = 9
    elif y == "K":
        y = 10
    elif y == "L":
        y = 11
    elif y == "M":
        y = 12
    elif y == "N":
        y = 13
    elif y == "O":
        y = 14
    else:
        return None  
    
    # ^Makes sure that this if one of the input is wrong, it wont do anything
    
    #print(x,y) --->To see if the positions are translated the way I wanted it to.
    
    
    if is_occupied(board,x,y) == False:   #if is_occupied at (A0) is False, then print stone and return True
        board[x][y] = stone               #at board [x][y], then place a stone 
        return True

    # ^We used the previous is_occupied function to see if the stone can be put or not.
    # ^If is_occupied is False, then the stone will be place at intended position.

    else:
        return False                      
    
    # ^Else, nothing will be printed.
    
#     return board                          
    
place_on_board(board, "●", ("3", "I"))


```

### 3.5 Printing the Board

### Implement code for 3.5 here



```python
#The purpose of this code is to take my 'raw' board and present it in a user-friendly way.
#The logic behind this is to take each row of my board, and for each element in my row, I would join it with the board characters. 

def print_board(board):
    
    
    all_letter_columns = "ABCDEFGHIJKLMNO"  
    
    # ^I needed to print the letters on my board so users can refer to it. I settled with just putting all my letters in
    #  string form instead of list ["A","B"...] because indexing works the same and I was more comfortable with this.
    
    
    for letter in range(len(board)):
        print (all_letter_columns[letter], " " ,end = " ")  
        
    # ^Here, for the length of board is used as a range. If it was 9, then it would go from 0-8. So 'letter' would be 0-8
    # ^The code in line 14, will be something like this, all_letter_columns[0] and will print the first character of the 
    #  all_letter_columns and put a space after it. It will then iterate in its range.
    #  But since I used the 'end' parameter, it will put all of the iteration in one line.
    

    cols = len(board)                        
    rows = 0 
    
    # ^I define this so I can later use it in my for loop 
    
    print("\n")                                   
    
    # ^Add space between board and letters.
    
    for row in board:                            
        print("---".join(board[rows]), " ", rows)
        rows+=1
        
    # ^This piece of code takes each row of my board and join each element in the row with a '---'
    # ^Since I already defined rows as 0, I will then increment with 1 to ensure  it goes to each row of my board.
        
        if rows < cols:                         
            print( "|   "* (cols))  
    # ^I had to put this code here because since this was in a for-loop, an extra row of '|'  was produced.
    #  and since the board has n numbers of '---' and has n-1 numbers of '|'.


print_board(board)
```

### 3.7 Check for the Winner

### Implement code for 3.7 here



```python
#The aim here is to find if theres any 5 stones of the same color in succession. 

#It checks each item in the board to see if there are any. For this, I used parts of my code from the previous functions 
# above to do so.

#The logic is to check if element1, element2, element3, element4, element5 is the same horizontally, vertically, and 
# diaganolly of both direction (/ and \). We also need to take in account for draw

#Part of this code was inspired by https://www.askpython.com/python/examples/connect-four-game

#A draw happens if all position on board is occupied and no 5 stones in succession was found.

    

def check_winner(board):  
    
    
   #Horizontal win

    for x in range(len(board)):              
        for y in range(len(board)):
            
    # ^The above code will tru each element in the board ex: [0][0],[0][1],[0][2]...
            
            if y + 4 < len(board):           
            
            # ^This condition here is to ensure that the y-index of the board is within the board-limits.

             if board[x][y] == board[x][y+1] == board[x][y+2] == board[x][y+3] == board[x][y+4]:   
            
            # ^This code here it to check for horizontal win. So the row will be the same but the column is not.
            # Ex: board[0][2] == board[0][3] == board[0][4] == board[0][5] == board[0][6] 
            
                    if board[x][y] == "●":
                        return "●"
                    elif board[x][y] == "○":
                        return "○"
            # ^So if "●" has 5 stones in succession horizontally, then they won. Else, "○"
            
                    
    #Vertical win
    
    for x in range(len(board)):              
        for y in range(len(board)):          
            
            if x + 4 < len(board): 
                
            # ^This condition here is to ensure that the x-index of the board is within the board-limits.
            
             if board[x][y] == board[x+1][y] == board[x+2][y] == board[x+3][y] == board[x+4][y]:
                    
            # ^This code here it to check for vertical win. So the column will be the same but the row is not.
            # Ex: board[0][2] == board[1][2] == board[2][2] == board[3][2] == board[4][2] 
                    
                    
                    if board[x][y] == "●":
                        return "●"
                    elif board[x][y] == "○":
                        return( "○")
            
                    
    #Diagonal(positive curve /)
    
    for x in range(len(board)):              
        for y in range(len(board)):          
            
            if x + 4 < len(board) and y-4 >=0:   
            
            # ^This condition here is to ensure that the x-index & y-index of the board is within the board-limits. y != <0
            
             if board[x][y] == board[x+1][y-1] == board[x+2][y-2] == board[x+3][y-3] == board[x+4][y-4]:  
                    
            # ^This code here it to check for diagnol win from of positive curve
            # Ex: board[1][6] == board[2][5] == board[3][4] == board[4][3] == board[5][2] 
            
                    if board[x][y] == "●":
                        return "●"
                    elif board[x][y] == "○":
                        return( "○")
                    
     #Diagonal(negative curve \ )
    
    for x in range(len(board)):              
        for y in range(len(board)):         
            
            if x + 4 < len(board) and y+4 < len(board): 
            
            # ^This condition here is to ensure that the x-index & y-index of the board is within the board-limits.
            
             if board[x][y] == board[x+1][y+1] == board[x+2][y+2] == board[x+3][y+3] == board[x+4][y+4]:
                    
            # ^This code here it to check for diagnol win from of negative curve
            # Ex: board[2][1] == board[3][2] == board[4][3] == board[5][4] == board[6][5] 
                    
                    if board[x][y] == "●":
                        return "●"
                    elif board[x][y] == "○":
                        return( "○")
    
    #Draw
    
    for x in range (len(board)):
        for y in range (len(board)):
            if board [x][y] == " ": 
                return None
        
    return "Draw"

    # ^Here if in theres no empty spaces in each row of the board, and none of the above winning condition
    #  was satisfied, then its a Draw.

    

check_winner(board)
```

```python
# The computer player in this game should put around a 3*3 radius of the stone the human player place.
# This one requires a bit of calculating.



def random_computer_player(board, player_move):
    
    
    all_letter_columns = "ABCDEFGHIJKLMNO"
    
    
    player_move_row = int(player_move[0])                                        
    player_move_col = all_letter_columns.index(player_move[1].upper())
    
    print(player_move_row,player_move_col)
    
    #The below code is for the instance where theres no valid computer moves, thus computer can put wherever
    
    available_moves = []                                                    
    
    for x in range(len(board)):
        for y in range(len(board)):
            if board[x][y] == " ":
                available_moves.append((x,y))
                
#     print(len(available_moves))
         
    #Below is for the valid computer moves.
    
    valid_comp_moves = []
    
    #If player_move is at (1,5), then I need to check the row above the player's move that is within radius
    
    for x in range(player_move_row-1, player_move_row):
        for y in range(player_move_col -1, player_move_col + 2):
            
            # ^The above code is based on calculation to ensure it is within radius
            
            if x >= 0 and x < len(board) and y >=0 and y < len(board):
                valid_comp_moves.append((x,y))
                
            # ^This code is important to ensure that the the x and y is within valid limits. 
            # ^If player_move is at (0,5), then no valid moves will be available.
            
            
    #The below code will check the row below the player's move that is within radius
    
    for x in range(player_move_row + 1, player_move_row + 2):
        for y in range(player_move_col -1, player_move_col + 2):
                        
            if x >= 0 and x < len(board) and y >=0 and y < len(board):
                valid_comp_moves.append((x,y))
                
            # ^This code is important to ensure that the the x and y is within valid limits.
            # ^If player_move is at (8,5), then no valid moves will be available.
            
    #The below code will check the left of the stone on the same row
    
    for x in range(player_move_row, player_move_row + 1):
        for y in range(player_move_col -1, player_move_col):
                        
            if x >= 0 and x < len(board) and y >=0 and y < len(board):
                valid_comp_moves.append((x,y))
                
    #The below code will check the right of the stone on the same row
                
    for x in range(player_move_row , player_move_row + 1):
        for y in range(player_move_col + 1, player_move_col + 2):
                        
            if x >= 0 and x < len(board) and y >=0 and y < len(board):
                valid_comp_moves.append((x,y))
            
    #print(valid_comp_moves)
    
#The below is not perfect. Please refer to test on why.
    
    if valid_comp_moves:
        
        computer_move = random.choice(valid_comp_moves)
        computer_row = computer_move[0]
        computer_col = computer_move[1]
#         return (computer_row, computer_col)
        
        if board[computer_row][computer_col]== " ":
            board[computer_row][computer_col] = "○"  
    
    else:
        computer_move = random.choice(available_moves)
        computer_row = computer_move[0]
        computer_col = computer_move[1]
#         return (computer_row, computer_col)
        
        if board[computer_row][computer_col]== " ":
            board[computer_row][computer_col] = "○"
            
        

random_computer_player(board, ("3","I"))
print_board(board)


```
